```python
import pickle
import tweepy
import csv
import pandas as pd
alltweets= pickle.load(open("./data/uber.p", "rb"))
alltweets_df=pd.DataFrame(alltweets,columns=['all_data','user','user_name','listed','followers','entities','retweets','reply','tweet'])
```

```python
formatted=[]
for index in range(len(alltweets_df)):
    source=alltweets_df['user'][index].screen_name
    tweet=alltweets_df['tweet'][index]
    date=alltweets_df['all_data'][index].created_at
    entities=alltweets_df['entities'][index]
    sourceid=alltweets_df['user'][index].id
    print(str(alltweets_df['reply'][index])!="None")
    if(str(alltweets_df['reply'][index])!="None"):
        type_tweet="Reply"
    elif(len(entities['user_mentions'])==0):
        type_tweet="Tweet"
    elif b'RT' in tweet[0:4]:
        type_tweet="Retweet"
    elif '@' in str(tweet):
        type_tweet="Mention"
    else:
        type_tweet=""

    mentions=[mention['screen_name'] for mention in alltweets_df['entities'][index]['user_mentions']]
    mentionid=[mention['id'] for mention in alltweets_df['entities'][index]['user_mentions']]
    for mentioned,mid in zip(mentions,mentionid):
        toadd=[source,sourceid,mentioned,mid,type_tweet,alltweets_df['all_data'][index],date]
        formatted.append(toadd)
    if(type_tweet=="Tweet"):
        toadd=[source,sourceid,source,sourceid,type_tweet,alltweets_df['all_data'][index],date]
        formatted.append(toadd)
#     print(source,mentioned)
```

```python
formatted=pd.DataFrame(formatted)
formatted.columns=['Source','Sourceid','Target','Targetid','type_of_content','all_data','created_at']
formatted
```
